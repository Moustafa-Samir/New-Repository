```python
import random
import pandas as pd
from tkinter import *
```

```python
'''This application creats table which all of rows sum & colimn sum equal to 
a given number "total" and table dimention also is a given number "matrix" ''' 

window = Tk()
window.title = "Create Table"
window.geometry("800x500")

lbl1=Label(window,text="Table dimention")
lbl2=Label(window,text="Sum of rows & columns")
lbl1.place(x=50,y=5)
lbl2.place(x=50,y=30)
lpl4=Label(window,text="Font size")
lpl4.place(x=50,y=50)
f_size=IntVar()
def spnbox():
   f_size.get()

sbox1 = Spinbox(window, from_ = 8 , to = 50,textvariable=f_size,command=spnbox())
sbox1.place(x=200,y=50)
matrix = IntVar()
total = IntVar()

def spnbox():
   f_size.get()

def Create_Table():
    '''This function takes (matrix & total) and creats a table
   (matrix) is the table dimention & (total) is the sum of each colmun & each row'''   
    
    global lbl3
    lbl3 = Label(window,font = f_size.get() )
    lbl3.place(x=50,y=80) 
    
    #matrix=int(matrix) 
    spread = int(total.get() / matrix.get() * 1.6)
    start = int(total.get() / matrix.get() / 2 )
       
    if  total.get() <  matrix.get() or matrix.get() < 1 :
       print("please enter logic combination")
       return 
       
    mydict = dict()                                                              # Create table 
    for i in range(matrix.get()):
      x = [random.randint(start,spread) for i in range(matrix.get())]          
      mydict[i] = x 
       
    for i in range(matrix.get()):                                                      # Adjust rows total
       mydict[matrix.get()-1][i] = 0
       for e in range(matrix.get() - 1):          
          mydict[matrix.get()-1][i] += int(mydict[e][i])
       mydict[matrix.get()-1][i] = total.get() - mydict[matrix.get()-1][i]  
       
    for i in range(matrix.get()):                                                      # Adjust columns total  
       mydict[i][matrix.get() - 1] = total.get() - sum(mydict[i]) + mydict[i][matrix.get() - 1]             
              
    mydict["Total"] = [0 for i in range(matrix.get())]                                 # Add total for rows  
    for i in range(matrix.get()):                                 
       for e in range(matrix.get()):
          mydict["Total"][i] += mydict[e][i]
           
    [mydict[i].append(sum(mydict[i])) for i in range(matrix.get())]                      #add total for columns

    mydict["Total"].append(sum(mydict["Total"]))                                   #add total for column total
    lbl3.config(text= pd.DataFrame(mydict),fg="red",font = f_size.get()) 
    btn2.config(state= "active")
    btn1.config(state="disabled")  

def delete():      
    btn1.config(state= "active")
    btn2.config(state="disabled")
    lbl3.destroy()
    lbl3.config(text= "",fg="red",font=f_size.get())
      
 
btn2= Button(window,text = "Delete Table",command = delete)
btn2.place(x=450,y=13)
btn1= Button(window,text = "Create Table",command = Create_Table)
btn1.place(x=350,y=13)
ent1= Entry(window, textvariable = matrix)
ent2= Entry(window, textvariable = total)
ent1.place(x=200,y=5)
ent2.place(x=200,y=30)
   
window.mainloop()
```
