```python
from tkinter import *
from tkinter import ttk
import math
"Calculator program"

window = Tk()
window.title("Calculator")
window.geometry("500x300")
window.maxsize(width=500,height=300)

x= IntVar()
y= IntVar()
z=StringVar()

lbl1 = Label(window,text = "Calculator",font = 70,height=2)
lbl1.pack()
lbl2 = Label(window,text = "Result",font = 50,height=2,background="black",foreground="white",width=25)
lbl2.pack()

def result():
    if Combobox1.get() == "+":
        lbl2["text"]= x.get() + y.get()
    elif Combobox1.get() == "-":
        lbl2["text"]= x.get() - y.get()    
    elif Combobox1.get() == "*":
        lbl2["text"]= x.get() * y.get()   
    elif Combobox1.get() == "/":
        try:
            lbl2["text"]= x.get() / y.get()     
        except:
            lbl2["text"]="You can\'t divide by zero"
    elif Combobox1.get() == "sin":
        lbl2["text"]= math.sin(x.get())
    elif Combobox1.get() == "cos":
        lbl2["text"]= math.cos(x.get()) 
    elif Combobox1.get() == "exp":
        lbl2["text"]= math.exp(x.get())   
    elif Combobox1.get() == "power":
        lbl2["text"]= math.pow(x.get() , y.get())  
    elif Combobox1.get() == "square root":
        lbl2["text"]= math.sqrt(x.get())      
        


btn1 = Button(text = "Calculate",width= 20,height=2,border = "7",command = result)
btn1.pack()
lbl3= Label(window,text="Enter first number")
lbl3.place(x=100,y=200)
ent1=Entry(window,width = 15,textvariable=x,font=15)
ent1.place(x=230,y=200)
lbl4= Label(window,text="Enter seconf number")
lbl4.place(x=100,y=230)
ent2=Entry(window,width = 15,textvariable=y,font=15)
ent2.place(x=230,y=230)

def check(): 
  if z.get() =="sin":   #in ["sin","cos","exp","square root","power"]:
     lbl2.pack_forget()

Combobox1= ttk.Combobox(window,values=["+","-","/","*","sin","cos","exp","square root","power"],textvariable = z,width=15,font=10)
Combobox1.place(x=230,y = 160)  
lbl5= Label(window,text="Choose operator")
lbl5.place(x=100,y=160)



window.mainloop()
```
